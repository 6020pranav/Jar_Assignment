# Part_1(A)

```python
import pandas as pd
```

```python
orders = pd.read_excel(r"C:\Users\PRANAV\OneDrive\Desktop\JAR\List_of_Orders_1.xlsx")
order_details = pd.read_excel(r"C:\Users\PRANAV\OneDrive\Desktop\JAR\Order_Details_1.xlsx")
```

```python
print(orders.head())
```

```python
print(order_details.head())
```

```python
order_id_col_orders = "Order ID"
order_id_col_details = "Order ID"
```

```python
merged_df = pd.merge(orders,order_details,left_on=order_id_col_orders,right_on=order_id_col_details,how="inner")
```

```python
category_col = "Category"
amount_col = "Amount"
```

```python
merged_df[amount_col] = pd.to_numeric(merged_df[amount_col], errors="coerce")
```

```python
print(merged_df.head())
```

```python
category_sales = merged_df.groupby(category_col)[amount_col].sum().reset_index()
category_sales = category_sales.sort_values(by=amount_col, ascending=False)
```

```python
print(category_sales)
```

# Part_1(B)

```python
profit_col = "Profit"
```

```python
print(merged_df.head())
```

```python
merged_df[amount_col] = pd.to_numeric(merged_df[amount_col], errors="coerce")
merged_df[profit_col] = pd.to_numeric(merged_df[profit_col], errors="coerce")
```

```python
summary = merged_df.groupby(category_col).agg(total_sales=(amount_col, "sum") avg_profit_per_order=(profit_col, "mean"), total_profit=(profit_col,"sum")).reset_index()
```

```python
print(summary)
```

```python
summary["profit_margin_pct"] = (summary["total_profit"] / summary["total_sales"]) * 100
summary = summary.sort_values(by="total_sales", ascending=False)
```

```python
print(summary)
```

# Part_1(c)

```python
top_by_sales = summary.sort_values(by="total_sales", ascending=False).head(3)
top_by_margin = summary.sort_values(by="profit_margin_pct", ascending=False).head(3)
under_by_margin = summary.sort_values(by="profit_margin_pct", ascending=True).head(3)
```
