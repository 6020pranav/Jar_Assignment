```python
import pandas as pd
```

```python
orders = pd.read_excel(r"C:\Users\PRANAV\OneDrive\Desktop\JAR\List_of_Orders_1.xlsx")
order_details = pd.read_excel(r"C:\Users\PRANAV\OneDrive\Desktop\JAR\Order_Details_1.xlsx")
```

```python
order_id_col_orders = "Order ID"
order_id_col_details = "Order ID"
```

```python
merged_df = pd.merge(orders, order_details, left_on=order_id_col_orders, right_on=order_id_col_details, how="inner")
```

```python
print(merged_df.head())
```

```python
merged_df["Amount"] = pd.to_numeric(merged_df["Amount"], errors="coerce")
merged_df["Profit"] = pd.to_numeric(merged_df["Profit"], errors="coerce")
```

```python
regional_summary = merged_df.groupby("State").agg(total_sales=("Amount", "sum"),avg_profit=("Profit", "mean"),order_count=(order_id_col_orders, "nunique")).reset_index()
```

```python
top_5_states = regional_summary.sort_values(by="order_count", ascending=False).head(5)
```
